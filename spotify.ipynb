```python
print("Original column names = ", df.columns)

df.columns= df.columns.str.replace(" ","_").str.lower()
print("Column names after conversion = ", df.columns)
```

Display the columns that have NaN values along with the count of NaN values. 

```python

nan_col = df.columns[df.isnull().any()]
for i in nan_col:
    print(i, df[i].isnull().sum())
```

```python
df_cleaned = df.drop(columns=['description', 'url_spotify', 'stream', 'url_youtube', 'uri'])

```

```python
threshold = 500
df_cleaned = df_cleaned.dropna(thresh=df_cleaned.shape[0] - threshold, axis=1)

```

```python
numeric_cols = ['danceability', 'energy', 'key', 'loudness', 'speechiness',
                'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']
for col in numeric_cols:
    if col in df_cleaned.columns:
        df_cleaned[col].fillna(df_cleaned[col].mean(), inplace=True)
```

```python
categorical_cols = ['url_youtube', 'title', 'channel', 'views', 'likes', 
                    'comments', 'licensed', 'official_video']
for col in categorical_cols:
    if col in df_cleaned.columns:
        df_cleaned[col].fillna(df_cleaned[col].mode()[0], inplace=True)
```
